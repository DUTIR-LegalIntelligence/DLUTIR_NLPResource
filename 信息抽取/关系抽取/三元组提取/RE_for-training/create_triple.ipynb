```python
import csv
import sys
import re
```

```python
pat_e1=r'\[E11\](.*?)\[E12\]'
pat_e2=r'\[E21\](.*?)\[E22\]'
with open('eval/sem_res.txt', "r", encoding="utf-8-sig") as f:
        reader = csv.reader(f, delimiter="\t")
        lines = []
        for line in reader:
            if sys.version_info[0] == 2:
                line = list(cell for cell in line)
            lines.append(line)
            
dic_triple={}
for i in range(len(lines)):
    #print(lines[i])
    if lines[i][3] not in dic_triple:
        dic_triple[lines[i][3]]=set()
#     print(lines[i][1])
    if lines[i][2]=='NA':
        continue
    e1=re.search(pat_e1,lines[i][1]).group(1)
    e2=re.search(pat_e2,lines[i][1]).group(1)
    triple=tuple((e1,lines[i][2],e2))
    dic_triple[lines[i][3]].add(triple)
    #print(triple)
    
print(dic_triple)
    
```

```python
def triple_evalute(dic_true,dic_pred):
    sum_pred=0.000000001
    sum_true=0.000000001
    sum_tp=0.000000001
    for key in dic_true:
        sum_true+=len(dic_true[key])
    for key in dic_pred:
        sum_pred+=len(dic_pred[key])
        if key in dic_true:
            for t in dic_pred[key]:
                t=list(t)
                if t in dic_true[key]:
                    sum_tp+=1
    precision=float(sum_tp/sum_pred)
    recall=float(sum_tp/sum_true)
    f1=float(2*precision*recall/(precision+recall))
    print(f1)
triple_evalute(dic_true,dic_triple)
```
